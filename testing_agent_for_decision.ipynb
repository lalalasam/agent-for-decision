```python
from qdrant_client.http.models import VectorParams, Distance, PayloadSchemaType, VectorsConfig

class HybridCollectionManager:
    def __init__(self, qdrant_url: str, qdrant_api_key: str):
        self.client = QdrantClient(url=qdrant_url, api_key=qdrant_api_key)

    def create_hybrid_collection(self, collection_name: str, dense_dim: int, sparse_dim: int):
        if self.client.collection_exists(collection_name):
            print(f"📁 Collection '{collection_name}' already exists.")
            return

        print(f"🚀 Creating hybrid-search collection: '{collection_name}'")

        vectors_config = {
            "dense": models.VectorParams(size=dense_dim, distance=Distance.COSINE),
            "sparse": models.VectorParams(size=sparse_dim, distance=Distance.DOT)
        }

        self.client.recreate_collection(
            collection_name=collection_name,
            vectors_config=vectors_config,
            on_disk_payload=True
        )

        self._create_payload_indexes(collection_name)
        print(f"✅ Hybrid collection '{collection_name}' created successfully.")

    def _create_payload_indexes(self, collection_name: str):
        self.client.create_payload_index(
            collection_name=collection_name,
            field_name="text",
            field_schema=PayloadSchemaType.TEXT
        )
        self.client.create_payload_index(
            collection_name=collection_name,
            field_name="keywords",
            field_schema=PayloadSchemaType.KEYWORD
        )
        self.client.create_payload_index(
            collection_name=collection_name,
            field_name="group_id",
            field_schema=PayloadSchemaType.KEYWORD
        )
```

```python
from qdrant_client import QdrantClient, models
from qdrant_client.http.models import Distance, PayloadSchemaType

class HybridCollectionManager:
    def __init__(self, qdrant_url: str, qdrant_api_key: str):
        self.client = QdrantClient(url=qdrant_url, api_key=qdrant_api_key)

    def create_hybrid_collection(self, collection_name: str, dense_dim: int, sparse_dim: int):
        if self.client.collection_exists(collection_name):
            print(f"📁 Collection '{collection_name}' already exists.")
            return

        print(f"🚀 Creating hybrid-search collection: '{collection_name}'")

        # Create collection with dense vectors
        self.client.create_collection(
            collection_name=collection_name,
            vectors_config=models.VectorParams(
                size=dense_dim,
                distance=Distance.COSINE,
                on_disk=True
            ),
            # Add sparse vectors configuration
            sparse_vectors_config={
                "sparse": models.SparseVectorParams(
                    index=models.SparseIndexParams()
                )
            },
            on_disk_payload=True
        )

        self._create_payload_indexes(collection_name)
        print(f"✅ Hybrid collection '{collection_name}' created successfully.")

    def _create_payload_indexes(self, collection_name: str):
        self.client.create_payload_index(
            collection_name=collection_name,
            field_name="text",
            field_schema=PayloadSchemaType.TEXT
        )
        self.client.create_payload_index(
            collection_name=collection_name,
            field_name="keywords",
            field_schema=PayloadSchemaType.KEYWORD
        )
        self.client.create_payload_index(
            collection_name=collection_name,
            field_name="group_id",
            field_schema=PayloadSchemaType.KEYWORD
        )
```

# Memory management system

```python
def add_memory_entry(
    data,
    user_id=None,
    agent_id=None,
    run_id=None,
    metadata=None,
    filters=None,
    prompt=None,
    collection_name=None,  # ⬅️ New argument
):
    metadata = metadata or {}
    filters = filters or {}

    if collection_name:
        memory.collection_name = collection_name  # ⬅️ Set target collection dynamically

    embeddings = memory.embedding_model.embed(data)

    if user_id:
        filters["user_id"] = metadata["user_id"] = user_id
    if agent_id:
        filters["agent_id"] = metadata["agent_id"] = agent_id
    if run_id:
        filters["run_id"] = metadata["run_id"] = run_id

    if not prompt:
        prompt = MEMORY_DEDUCTION_PROMPT.format(user_input=data, metadata=metadata)

    # Ask LLM to extract memories
    extracted_memories = memory.llm.generate_response(
        messages=[
            {"role": "system", "content": "You are an expert at deducing facts, preferences, and memories."},
            {"role": "user", "content": prompt},
        ]
    )

    # Search for similar existing memories
    existing = memory.vector_store.search(
        name=memory.collection_name,
        query=embeddings,
        limit=5,
        filters=filters,
    )
    existing_memories = [
        MemoryItem(id=mem.id, score=mem.score, metadata=mem.payload, text=mem.payload.get("data", ""))
        for mem in existing
    ]

    # Prepare update messages
    update_messages = get_update_memory_messages(
        [m.model_dump(include={"id", "text", "score"}) for m in existing_memories],
        extracted_memories
    )

    # Tool-enabled response
    response = memory.llm.generate_response(messages=update_messages, tools=[ADD_MEMORY_TOOL, UPDATE_MEMORY_TOOL, DELETE_MEMORY_TOOL])
    tool_calls = response.get("tool_calls", [])

    results = []

    if tool_calls:
        tool_map = {
            "add_memory": memory._create_memory_tool,
            "update_memory": memory._update_memory_tool,
            "delete_memory": memory._delete_memory_tool,
        }
        for call in tool_calls:
            name = call["name"]
            args = call["arguments"]
            logging.info(f"[Mem0] Executing: {name} with args: {args}")

            if name in ["add_memory", "update_memory"]:
                args["metadata"] = metadata

            result_id = tool_map[name](**args)
            results.append({
                "id": result_id,
                "event": name.replace("_memory", ""),
                "data": args.get("data"),
            })

            capture_event("mem0.add.function_call", memory, {
                "memory_id": result_id,
                "function_name": name,
            })

    capture_event("mem0.add", memory)

    return results
```

```python
###Wrap add_memory_entry() as an Agent Tool

def add_memory_tool(query: str) -> str:
    user_id = os.getenv("ACTIVE_USER_ID", "default_user")
    agent_id = os.getenv("ACTIVE_AGENT_ID", "portfo_agent")
    collection = os.getenv("ACTIVE_COLLECTION", "agent_memories")
    run_id = f"req-{uuid.uuid4().hex[:8]}"

    memory_results = add_memory_entry(
        data=query,
        user_id=user_id,
        agent_id=agent_id,
        run_id=run_id,
        collection_name=collection
    )

    extracted_texts = [m.get("data", "") for m in memory_results if m.get("data")]
    return "🧠 Memory deduced:\n" + "\n".join(extracted_texts) if extracted_texts else "🤔 No memory context extracted."
```
